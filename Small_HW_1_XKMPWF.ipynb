```python
#in this cell I create the arrays for the histograms to plot in the next cells
#the cv2 package computes histograms automaticly with the calcHist method
blue_histogram1 = cv2.calcHist([img1], [0], None, [256], [0, 256])
green_histogram1 = cv2.calcHist([img1], [1], None, [256], [0, 256])
red_histogram1 = cv2.calcHist([img1], [2], None, [256], [0, 256]) 

blue_histogram2 = cv2.calcHist([img2], [0], None, [256], [0, 256])
green_histogram2 = cv2.calcHist([img2], [1], None, [256], [0, 256])
red_histogram2 = cv2.calcHist([img2], [2], None, [256], [0, 256]) 

blue_histogram3 = cv2.calcHist([img3], [0], None, [256], [0, 256])
green_histogram3 = cv2.calcHist([img3], [1], None, [256], [0, 256])
red_histogram3 = cv2.calcHist([img3], [2], None, [256], [0, 256]) 

blue_histogram4 = cv2.calcHist([img4], [0], None, [256], [0, 256])
green_histogram4 = cv2.calcHist([img4], [1], None, [256], [0, 256])
red_histogram4 = cv2.calcHist([img4], [2], None, [256], [0, 256]) 
 
blue_histogram5 = cv2.calcHist([img5], [0], None, [256], [0, 256])
green_histogram5 = cv2.calcHist([img5], [1], None, [256], [0, 256])
red_histogram5 = cv2.calcHist([img5], [2], None, [256], [0, 256]) 
```

```python
#to standardize the sets, we need to substract the mean and then divide by the std
#with more images using a for loop would have been more efficient, but this will work for now
img1_B_norm = (img1[:,:,0]-np.mean(img1[:,:,0]))/np.std(img1[:,:,0])
img1_G_norm = (img1[:,:,1]-np.mean(img1[:,:,1]))/np.std(img1[:,:,1])
img1_R_norm = (img1[:,:,2]-np.mean(img1[:,:,2]))/np.std(img1[:,:,2])

img2_B_norm = (img2[:,:,0]-np.mean(img2[:,:,0]))/np.std(img2[:,:,0])
img2_G_norm = (img2[:,:,1]-np.mean(img2[:,:,1]))/np.std(img2[:,:,1])
img2_R_norm = (img2[:,:,2]-np.mean(img2[:,:,2]))/np.std(img2[:,:,2])

img3_B_norm = (img3[:,:,0]-np.mean(img3[:,:,0]))/np.std(img3[:,:,0])
img3_G_norm = (img3[:,:,1]-np.mean(img3[:,:,1]))/np.std(img3[:,:,1])
img3_R_norm = (img3[:,:,2]-np.mean(img3[:,:,2]))/np.std(img3[:,:,2])

img4_B_norm = (img4[:,:,0]-np.mean(img4[:,:,0]))/np.std(img4[:,:,0])
img4_G_norm = (img4[:,:,1]-np.mean(img4[:,:,1]))/np.std(img4[:,:,1])
img4_R_norm = (img4[:,:,2]-np.mean(img4[:,:,2]))/np.std(img4[:,:,2])

img5_B_norm = (img5[:,:,0]-np.mean(img5[:,:,0]))/np.std(img5[:,:,0])
img5_G_norm = (img5[:,:,1]-np.mean(img5[:,:,1]))/np.std(img5[:,:,1])
img5_R_norm = (img5[:,:,2]-np.mean(img5[:,:,2]))/np.std(img5[:,:,2])
```

```python
d_of_norm_chs = {'img1': [img1_B_norm, img1_G_norm, img1_R_norm],
                 'img2': [img2_B_norm, img2_G_norm, img2_R_norm],
                 'img3': [img3_B_norm, img3_G_norm, img3_R_norm],
                 'img4': [img4_B_norm, img4_G_norm, img4_R_norm],
                 'img5': [img5_B_norm, img5_G_norm, img5_R_norm]}
#here I make a temporary dict to access the data of the images more efficiently and use it right after, to print what we need
for arr in d_of_norm_chs.keys():
    print(f'[{arr}, expected value = {np.mean(d_of_norm_chs[arr][0]):.1f}, std = {np.std(d_of_norm_chs[arr][0]):.1f}]',
          f'[{arr}, expected value = {np.mean(d_of_norm_chs[arr][1]):.1f}, std = {np.std(d_of_norm_chs[arr][1]):.1f}]'
          f'[{arr}, expected value = {np.mean(d_of_norm_chs[arr][2]):.1f}, std = {np.std(d_of_norm_chs[arr][2]):.1f}]')
```

The standardizing worked, all expected values (empirical means) are around 0 and standard deviations are 1. I could have used the standardscaler from sklearn, but it's easy to compute anyway.

```python
source = driver.page_source #using soup to format the raw code into a more convenient shape
soup=BeautifulSoup(source)
soup.prettify
```

```python
text = soup.get_text() #this simple method gets us all the text from the website
```

```python
def func(value): #Before printing, I remove the newline characters, because there was a lot of them
    return ''.join(value.splitlines())
print(func(text))
```

```python
simpler_text = func(text) #removing newline characters
even_simpler_text = simpler_text.replace(" ", "") #removing whitespace characters
THE_SIMPLEST_TEXT_TO_EVER_EXIST = even_simpler_text.lower() #changing uppercase letters to lowercase, to count the letters
#sorry for dramatizing
```

```python
from collections import Counter 
counter = Counter(THE_SIMPLEST_TEXT_TO_EVER_EXIST) 
#this function counts the distinctive characters in a string and puts the data in a dict
```

```python
the_dict_the_world_needs = {} #the counter has non alpha characters aswell, let's remove them
for key in counter.keys():
    if key.isalpha():
        the_dict_the_world_needs[key] = counter[key]
    else:
        continue
```
